```python
from tensorflow.keras.utils import image_dataset_from_directory as ifd
 
train = ifd(dir,validation_split = 0.2,
           subset = "training",
           seed = 123,
           image_size = (224,224),
           batch_size = 32)
```

```python
val_data = ifd(dir,validation_split = 0.3,
              subset = "validation",
              seed = 123,
              image_size = (224,224),#height, width
               batch_size = 32
              )
```

```python
from tensorflow.keras import layers

model = tf.keras.Sequential([
    layers.Rescaling(1./255, input_shape = (224,224,3)),
    layers.Conv2D(26, kernel_size =(5, 5), strides =(1, 1),activation ='relu'),
    layers.MaxPooling2D(),
    layers.Dropout(0.5),
    layers.Flatten(),
    layers.Dense(128,activation = "relu"),
    layers.Dense(2,activation="sigmoid")
])

model.compile(optimizer = "Adam", loss = tf.keras.losses.SparseCategoricalCrossentropy(),
              metrics = ["accuracy"])
epochs = 10
history = model.fit(train,epochs = epochs,validation_data = val_data,batch_size = 32)
```
